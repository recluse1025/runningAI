```python
#路線篩選函式
def road(type3,level,city):
    #mysql連線
    import mysql.connector
    db = mysql.connector.connect(user='root', password='1234', host='192.168.200.149',port='3306', database="final",
                                 charset = 'utf8')
    #mysql指令容器
    cursor = db.cursor()    
    
    #篩選登山或跑步
    type3_dict = {"hill":" > 150 ","run":" < 150 "}
    x = type3_dict.get(type3)
    
    #篩選難易度
    level_dict ={"strong":" '高' ","medium":" '中' ","weak":" '低' "}
    y = level_dict.get(level)

    #篩選地區
    city_dict = {"TPE":"%台北%", "NTPC":"%新北%", "KEL":"%基隆%", "TYN":"%桃園%", "HSZ":"%新竹%",
                "ZMI":"%苗栗%", "ILA":"%宜蘭%", "TXG":"%台中%", "CHW":"%彰化%", "NTC":"%南投%", 
                 "YUN":"%雲林%", "CYI":"%嘉義%", "TNN":"%台南%", "KHH":"%高雄%", "PIF":"%屏東%",
                "PEH":"%澎湖%"}
    z = city_dict.get(city)
    
    #sql篩選
    cursor.execute("SELECT final.mountain_basic.url, final.mountain_basic.標題, final.mountain_basic.評分,\
    final.mountain_basic.里程, final.mountain_basic.海拔高度, final.mountain_basic.難易度, final.mountain_city.所在縣市 \
    FROM final.mountain_basic \
    LEFT JOIN final.mountain_city ON final.mountain_basic.標題 = final.mountain_city.標題 \
    WHERE final.mountain_basic.高度落差"+ x +"\
    AND final.mountain_city.所在縣市 LIKE '"+ z +"' \
    AND final.mountain_basic.難易度 LIKE" + y)  
    
    #建立字典清單
    res_list = []
    for x in cursor:
        #url
        url = x[0]
        #標題
        name = x[1]
        #評分
        score = x[2]
        #里程
        mileage = x[3]
        #海拔高度
        high = x[4]
        #難易
        level = x[5]
        #縣市
        city = x[6]
        res = {'url': url, 'name': name, 'score':score, 'mileage':mileage,'high':high, 'level':level, 'city':city}
        res_list.append(res)
    db.close()
    return res_list

#使用函式
road("hill","weak","NTPC")
```
