```python

import datetime
import time
import random
a1=(2019,11,1,6,0,0,0,0,0)       #设置开始日期时间元组（1976-01-01 00：00：00）
a2=(2019,11,20,12,59,59,0,0,0)  #设置结束日期时间元组（1990-12-31 23：59：59）
a3=(2019,11,1,6,0,0,0,0,0)
a4=(2019,11,1,12,59,59,0,0,0)
start=time.mktime(a1)  #生成开始时间戳
end=time.mktime(a2)   #生成结束时间戳
start2=time.mktime(a3)
end2=time.mktime(a4)
#随机生成10个日期字符串

for i in range(495):

    bpm = random.randint(120,160)
    type5 = random.choice(['run','walk'])
    if type5 == 'run':
        pace_second = random.randint(300,480)
        pace = round(pace_second / 60, 2)  # random.randint(5,8)
    else:
        pace_second = random.randint(540,900)
        pace = round(pace_second / 60, 2)  # random.randint(5,8)
    distance = random.choice([5,10,14,21,42])
    total_time = pace_second*distance
    user_id = random.randint(800,820)
    


    t=random.randint(start,end)  #在开始和结束时间戳中随机取出一个
    y=random.randint(start2,end2)
    date_touple=time.localtime(t)     #将时间戳生成时间元组
    date_touple2=time.localtime(y)
    date=time.strftime("%Y-%m-%d ",date_touple) #将时间元组转成格式化字符串（1976-05-21）
    date2=time.strftime("%H:%M:%S",date_touple2)
    date3=date+date2
    date4=datetime.datetime.strptime(date3,'%Y-%m-%d %H:%M:%S')


    td = datetime.timedelta(seconds=total_time)
    date5 = date4+td    
#     print(user_id,bpm,pace,distance,total_time,date4,date5,type5)

    second = total_time % 60
    minute = (total_time // 60) % 60
    hour = (total_time // 60) // 60
    total_time = "{}:{}:{}".format(int(hour), int(minute), int(second))
    
    dic = {"user_id":user_id,
           "bpm":bpm,
           "pace":pace,
           "distance":distance,
           "total_time":total_time,
           "start_time":date4,
           "end_time":date5,
           "type5":type5}

    user_id = str(dic.get("user_id"))
    bpm = str(dic.get("bpm"))
    pace = str(dic.get("pace"))
    distance = str(dic.get("distance"))
    total_time = str(dic.get("total_time"))
    start_time = str(dic.get("start_time"))
    end_time = str(dic.get("end_time"))
    type5 = str(dic.get("type5"))
    #print(dic['bpm'])
    print(type(user_id), bpm, pace, distance, total_time, start_time, end_time, type5)
    import pyodbc
    # Connect to LongDB
    cnxn=pyodbc.connect("DSN=SpliceODBC64")

    # Call sql command
    cursor=cnxn.cursor()

    # Run sql command
    cursor.execute("INSERT INTO SPLICE.HISTORY\
    (USER_ID, BPM, PACE, DISTANCE, TOTAL_TIME, START_TIME, END_TIME, TYPE5)\
    VALUES('"+user_id+"', '"+bpm+"', '"+pace+"', '"+distance+"', '"+total_time+"', '"+start_time+"', '"+end_time+"', '"+type5+"')")
    
    cnxn.commit()
```
